# Text Transformation
Data Cleaning (Removing unimportant data/ Stopwords/ Stemming)

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

nltk.download('stopwords')

# Initialize SnowballStemmer and stopwords
ss = SnowballStemmer('english')
stop = set(stopwords.words('english'))

def cleanText(text):
    text = re.sub('[^a-zA-Z]', ' ', text)  
    text = text.lower()  # Convert text to lowercase
    text = text.split()  # Split text into words
    words = [ss.stem(word) for word in text if word not in stop]  
    return " ".join(words)

data["text"] = data["text"].apply(cleanText)
data.head(n=10)
```

```python

non_integer_spam = data[~data["spam"].astype(str).str.isdigit()]

# Remove rows with non-integer values from the DataFrame
data = data[data["spam"].astype(str).str.isdigit()]

# Convert the values in the "spam" column to integers
data['spam'] = data['spam'].astype(int)

# Filter the DataFrame to extract spam and ham messages
spam_messages = data[data["spam"] == 1]["text"]
ham_messages = data[data["spam"] == 0]["text"]
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Initialize empty lists
spam_words = []
ham_words = []

for index, row in data.iterrows():
    words = word_tokenize(row['text'])  
    if row['spam'] ==1:  # Spam message
        spam_words.extend(words)  
    else:  # Ham message
        ham_words.extend(words)  
```

# Model Building :

```python
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import make_pipeline

# Instantiate CountVectorizer
cv = CountVectorizer()

# Fit and transform the text data
X = cv.fit_transform(data["text"])

# Impute missing values
imputer = SimpleImputer(strategy='most_frequent')
X_imputed = imputer.fit_transform(X)

# Create and train Multinomial Naive Bayes classifier
clf = MultinomialNB()
clf.fit(X_imputed, data["spam"])


```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

# Impute missing values in X
imputer = SimpleImputer(strategy='most_frequent')
X_imputed = imputer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.20, random_state=0)

# Train the model using MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

# Make predictions
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import fbeta_score

print(accuracy_score(y_test,y_pred))
print(fbeta_score(y_test,y_pred,beta =0.5))
```

```python
# evaluation matrcies

print (classification_report(y_test,y_pred))
```

```python
 # Model Evaluation | Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(dpi=100)
sns.heatmap(cm, annot=True)
plt.title("Confusion matrix")
plt.show()
```
